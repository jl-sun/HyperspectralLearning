```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import os

DATASET_DIR = "../_dataset/"
OUTPUT_DIR = "preview/"
if not os.path.exists(OUTPUT_DIR):
    os.mkdir(OUTPUT_DIR)
```

```python
FILE_NAME = "Indian_pines_corrected"
all_img = sio.loadmat(DATASET_DIR + FILE_NAME + ".mat")[FILE_NAME.lower()]
```

```python
import matplotlib.cm as colormap
for i in range(0, 103):
    if os.path.exists('{}{}.png'.format(OUTPUT_DIR, i)): break
    img = all_img[:, :, i]
    plt.imshow(img, cmap=colormap.get_cmap('gray'))
    plt.savefig('{}{}.png'.format(OUTPUT_DIR, i))
```

```python
print("Total max:", np.max(all_img))
print("Total min:", np.min(all_img))
max_arr = []
min_arr = []
for i in range(0, 103):
    img = all_img[:, :, i]
    max_arr.append(np.max(img))
    min_arr.append(np.min(img))
plt.plot(min_arr, 'r-')
plt.plot(max_arr, 'b-')
# plt.plot(np.array(max_arr)-np.array(min_arr), 'g--')
```
